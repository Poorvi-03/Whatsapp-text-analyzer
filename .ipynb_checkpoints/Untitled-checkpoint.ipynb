# Preprocess data

## Seperating msgs

```python


pattern ='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[a,p]m\s-\s'
messages = re.split(pattern,data)[1:]  #['Rahul Jamia: #Senior jawab do🤣']
dates= re.findall(pattern,data)
```

```python
messages
```

```python
dates
```

## creating dataframe

```python
df = pd.DataFrame({'user_messages': messages,
                       'message_date': dates})
```

```python
df.head()
```

```python
# creating function for seperating date and time
def gettimeanddate(string):   # line 25
    string = string.split(',')
    date,time = string[0],string[1]
    time = time.split('-')
    time = time[0].strip()
    return  date+ " " +time
```

```python
df['message_date']=df['message_date'].apply(lambda text: gettimeanddate(text))
```

```python
df.rename(columns={'message_date':'date'},inplace =True)
```

```python
df['user_messages']
```

```python
users=[]
messages =[]

for message in df['user_messages']:

    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])

    else:
        users.append('Group Notification')
        messages.append(entry[0])
```
